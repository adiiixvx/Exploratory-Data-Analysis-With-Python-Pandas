## Cleaning the data

```python
#only want USA races ,50km ,2020
df.dtypes
```

```python
#getiing the data with 50km/50mi
df[df["Event distance/length"].isin(["50km","50mi"])]
```

```python
#now data only with year 2020
df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020)]
```

```python
#event name has country name , so filtering with  it
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all filters together
df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df.shape
```

```python
#removing word USA from event name ,already all the data is only for USA
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#cleaning the athlete age
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete perf.
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get (0)
```

```python
df2.head(8)
```

```python
#droping unwanted columns
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

```python
df2.head(8)
```

```python
#clean upnull values
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
#check for duplicates
df2[df2.duplicated() == True]
```

```python
#reset index
df2.reset_index(drop = True)
```

```python
#fix types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head(8)
```

```python
#rename columns
df2 = df2.rename(columns = { 'Year of event':'year' ,
                 'Event dates':'race_day' ,
                 'Event name':'race_name' ,
                 'Event distance/length':'race_lenght' ,
                 'Event number of finishers': 'race_number_of_finishers',
                 'Athlete performance':'athlete_performace' ,
                 'Athlete gender':'athlete_gender' ,
                 'Athlete average speed':'athlete_average_speed' ,
                 'Athlete ID':'athlete_id' ,
                })
```

```python
df2.head()
```

```python
#reorder ther columns
df3 = df2[['race_day','race_name','race_lenght','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performace','athlete_average_speed']]
```

```python
df3.head(5)
```

```python
#vis.
sns.histplot(df3['race_lenght'])
```

```python
sns.histplot(df3, x = 'race_lenght', hue = 'athlete_gender')
```

```python
sns.displot(df3[df3['race_lenght'] == '50mi']['athlete_average_speed'] )
```

## After cleaning the data here we are going to get some answers for the que

```python
#q1 diff in speed for the 50k,50mi male to female
df3.groupby(['race_lenght','athlete_gender']) ['athlete_average_speed'].mean()
```

```python
#q2 what age groups are the best in the 50mi race (20+ races min)
df3.query('race_lenght == "50mi"').groupby(['athlete_age']) ['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>19')
```
